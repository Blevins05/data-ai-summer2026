También esta el indexado condicional, que es lo más parecido a las queries de SQL.

El **indexado condicional** de una columna también es algo muy útil. La sintaxix es `df[P(df['Name'])]`, donde  `P` es alguna condición lógica que es comprobada para cada elemento de la columna `Name`. El resultado de dicho indexado es el Dataframe que consta solo de las filas que satisfacen la condición `P` en la columna `Name`.


Por ejemplo, todos los partidos en los que se han anotado +3'5 goles por parte de algún equipo

````python
data[data['gf'] >= 4].shape
````

Otro ejemplo, cual es la posesión promedio de los partidos que terminan 0 a 0, para equipos que juegan en casa

Hay 233 partidos en los que ha habido más de 3 goles por parte de 1 de los 2 equipos

Apunte: En pandas no se puede usar 'and' en indexado condicional, se usa & y cada condición se envuelve en ().


````python
data[(data['gf'] == 0) & (data['ga'] == 0) & (data['venue'] == 'Home')]['poss'].mean()
````

La posesión seria de casi el 51% cuando se juega en casa.

También se puede aplicar a filas, solo hay que utilizar `axis=1`.
Las funciones lambda suelen ser muy útiles en estos casos.

Por ejemplo, partidos donde el resultado empieza por W que son Victorias (no es muy útil pero como ejemplo sirve, ya que con hacer data[data['result'] == 'W'] lo tienes

Otro ejemplo:
 Mediante una lambda, y aplicando a filas, podemos clasificar los resultados en puntos:

- W: 3 pts

- D: 1 pts

- L: 0 pts


````python
data['points'] = data['result'].apply(lambda x: 3 if x == 'W' else 1 if x == 'D' else 0)
````

Además, en este ejemplo hemos creado una nueva entrada en nuestro dataframe, añadiendo el descriptor points o puntos como nueva columna

````python
data.shape
````

Ahora tendríamos 30 descriptores

````python
data['points'].describe()
````

El método `map` se puede utilizar para reemplazar valores en una columna pasándole un diccionario de la forma `{old_value: new_value}` como argumento:

Convierto de W/D/L a Victoria/Empate/Derrota

````python
data['result'] = data['result'].map({'W': 'Victoria', 'D': 'Empate', 'L' : 'Derrota'})
````

````python
data['result'].value_counts()
````

Agrupaciones. Para construir grupos de datos en Pandas, utilizamos generalmente una construcción como la siguiente.

```python
df.groupby(by=grouping_columns)[columns_to_show].function()
```

1. Primero, el método `groupby` divide `grouping_columns` por sus valores, que se convierten en un nuevo índice en el dataframe resultante.
2. Seguidamente, las columnas de interés se seleccionan (`columns_to_show`). Si no se incluye `columns_to_show` se muestran todas las clausulas que no hayan sido agrupadas.
3. Finalmente, una o varias funciones se aplican para obtener los grupos por las columnas seleccionadas.


Por ejemplo, las siguientes agrupaciones son bastante útiles

Media de goles por equipo:

````python
data.groupby('team')['gf'].mean()
````

El Barcelona es el equipo que más goles ha promediado en estas 5 últimas temporadas, con 2.17 goles por partido. Seguido por el Real Madrid, con 2.01

El Real Oviedo es el equipo con menos goles promediado por encuentro que ha pasado por LaLiga en estos últimos 5 años.

Victorias, derrotas y empates por equipo

````python
data.groupby(['team', 'result']).size()
````

Puntos totales de cada equipo por temporada.

````python
data.groupby(['team', 'season'])['points'].sum()
````

Si quisieramos mostrar este último ejemplo completo, podemos utilizar `reset_index` que convierte el índice jerárquico en columnas normales y muestra todo como un DataFrame limpio.

````python
data.groupby(['team', 'season'])['points'].sum().reset_index()
````

Estadísticas de posesión y xG por resultado

````python
data.groupby('result')[['poss', 'xg', 'sh']].mean()
````

Se puede hacer algo parecido, pero en este caso pasando una lista de funciones a `agg()`:

````python
data.groupby('team')['gf'].agg(['mean', 'sum', 'max', 'min'])
````

Las tablas resumen en un dataframe nos sirven para ver como las muestras del mismo se distribuyen en el contexto de 2 variables.

Para esto se utiliza `crosstab()` junto a las 2 columnas que se quiere estudiar

Por ejemplo, cuantas veces cada equipo gana/empata/pierde

````python

pd.crosstab(data['team'], data['result'])
````
